این نوت‌بوک شامل پیاده‌سازی و شبیه‌سازی مدل‌های مختلف CRC است.

## محاسبه CRC برای مدل‌های مختلف

```

def compute_crc(data, poly, init, xor_out, width):
    crc = init
    for byte in data:
        crc ^= (byte << (width - 8))
        for _ in range(8):
            if crc & (1 << (width - 1)):
                crc = ((crc << 1) ^ poly) & ((1 << width) - 1)
            else:
                crc = (crc << 1) & ((1 << width) - 1)
    return crc ^ xor_out

models = {
    "CRC-8":  {"poly":0x07,   "init":0x00, "xor":0x00, "width":8},
    "CRC-16": {"poly":0x1021, "init":0xFFFF, "xor":0x0000, "width":16},
    "CRC-32": {"poly":0x04C11DB7, "init":0xFFFFFFFF, "xor":0xFFFFFFFF, "width":32},
}

data = b"Hello CRC"
results = {}

for name, m in models.items():
    results[name] = hex(compute_crc(data, m["poly"], m["init"], m["xor"], m["width"]))

results
```
